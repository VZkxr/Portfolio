## Simulation of planet

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd  
import seaborn as sns
```

```python
earth = {
    'África': {
        'initial_population': 1.4,
        'share women': .51
    },
    'Europa': {
        'initial_population': .749,
        'share women': .54
    },
    'América':{
        'initial_population': 1,
        'share women': .49
    },
    'Asia': {
        'initial_population': 4.7,
        'share women': .47
    },
    'Oceanía': {
        'initial_population': .046,
        'share women': .56
    }
}
```

### Continent

```python
class Continent:
    def __init__(self,inputs):
        self.name = inputs['name']
        self.initial_population = inputs['initial_population']
        self.share_women = inputs['share women']
```

```python
a = Continent({
    'name': 'América',
    'initial_population': earth['América']['initial_population'],
    'share women': earth['América']['share women']
})
```

```python
for continent in earth.keys():
    earth[continent]['class'] = Continent({
        'name': continent,
        'initial_population': earth[continent]['initial_population'],
        'share women': earth[continent]['share women']
    })
```

### Add growth rate

```python
for continent in earth.keys():
    earth[continent]['class'].growth_rate = earth[continent]['class'].share_women/5
```

### Forecast growth

```python
years2model = 50
time = range(years2model+1)

for continent in earth.keys():
    forecast_pop = []
    for t in time:
        y = earth[continent]['class'].initial_population * np.exp(t*earth[continent]['class'].growth_rate)
        forecast_pop.append(y)
    earth[continent]['class'].forecast = forecast_pop
```

```python
plt.figure(figsize = (10,8))

for continent in earth.keys():
    plt.plot(range(len(earth[continent]['class'].forecast)),earth[continent]['class'].forecast,label=continent)
plt.legend()
plt.show()
```
