### Librerías y variables

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
INGRESOS = 25000
PERIODO = 12

gasto_fijo = 5000
gasto_variable = (INGRESOS-gasto_fijo)/4

EGRESOS = gasto_variable + gasto_fijo; EGRESOS
```

### Activación

```python
activo = False
if INGRESOS > EGRESOS:      
    activo = True
else: 
    print("Advertencia: estás gastando más de lo que deberías")
```

```python
if activo:
    ganancia_anual = INGRESOS*PERIODO
    gasto_anual = EGRESOS*PERIODO
    
    ahorro = ganancia_anual - gasto_anual
    print(f"Estás ahorrando ${ahorro} anuales")
```

### Meta: Tener casa propia a los 30 años

```python
#COSTOS DEL TERRENO Y CONSTRUCCIÓN
COSTO_TERRENO = {
    'CDMX': 28000,
    'MTY': 21000,
    'GDL': 20000
}

COSTO_CONST = {
    'Residencial': (7000 + 11000)/2,
    'Económico': (5500 + 7500)/2 
}

#Variables de decisión
dimension_casa = 90
ciudad = 'CDMX'
modelo = 'Residencial'
```

```python
#CÁLCULO DEL COSTO
ct = COSTO_TERRENO[ciudad]*dimension_casa
cc = COSTO_CONST[modelo]*dimension_casa

COSTO_TOTAL = ct + cc
print(f'Costo de mi casa deseada: ${COSTO_TOTAL}')
```

### Estrategia

```python
#Estrategia estacionaria vs estrategia con cetes
T = 15
simu_cete = [ahorro]
tasa = 11.1/100
for i in range(T):
    y = ahorro + simu_cete[i]*(1+tasa)
    simu_cete.append(y)

tiempo = np.arange(T+1)
simu_est = ahorro*tiempo

plt.figure(figsize = (10,6))
plt.title('Estrategia para lograr mi meta')

plt.plot(tiempo,simu_est,label='Estrategia estacionaria')

plt.plot(6,COSTO_TOTAL,'ro',label='Costo casa')

plt.plot(tiempo,simu_cete,label='Estrategia cete')

plt.legend()
plt.show()
```
