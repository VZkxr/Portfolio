```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Condiciones iniciales
inputs = {
    'initial_wage': 25000,
    'annual_wage_growth': 0.03,

    'initial_basket_price': 4000,
    'annual_inflation': 0.047,

    'initial_rent_cost': 5000,
    'annual_rent_cost_growth': 0.05, 

    'quality_basis': 1000,
    'annual_quality_growth': 1000
}
```

```python
def model_wage(time,inputs):
    s = inputs['initial_wage']
    f = inputs['annual_wage_growth']

    y = s*(1+f)**(time/12)
    return y 

def model_basket(time,inputs):
    b = inputs['initial_basket_price']
    f = inputs['annual_inflation']

    y = b*(1+f)**(time/12)
    return y 

def model_rent(time,inputs):
    b = inputs['initial_rent_cost']
    f = inputs['annual_rent_cost_growth']

    y = b*(1+f)**(time/12)
    return y 
def model_quality(time,inputs):
    a = inputs['quality_basis']
    c = inputs['annual_quality_growth']

    y = a+(c*time/12)
    return y 

def model_expenses(time,inputs):
    y = model_basket(time,inputs) + model_rent(time,inputs) + model_quality(time,inputs)
    return y
```

```python
TIME = np.arange(0,12*51,12)
WAGE = model_wage(TIME,inputs)
BASKET = model_basket(TIME,inputs)
RENT = model_rent(TIME,inputs)
EXPENSES = model_expenses(TIME,inputs)
QUALITY = model_quality(TIME,inputs)
```

```python
#Visualizacin
plt.figure(figsize = (8,6))
plt.title('Wage and Basket evolution')
plt.plot(TIME,WAGE,label='wage')
plt.plot(TIME,BASKET,'--',label='basket',)
plt.plot(TIME,RENT,'--',label='rent')
plt.plot(TIME,QUALITY,'--',label='quality')
plt.plot(TIME,EXPENSES,label='expenses')
plt.xlabel('Time - Months')
plt.ylabel('MXN ($)')
plt.legend()
plt.show()
```
